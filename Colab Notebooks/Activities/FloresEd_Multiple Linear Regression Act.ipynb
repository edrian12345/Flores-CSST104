# 1.   Import Libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# **2. Inserting Dataset:**


```python
data = pd.read_csv('/content/Lesson_4_student_grades_dataset.csv')
print(data)
```

# **3. Visualizing Data:**

```python
x = data[['Attendance_Percentage']]
y = data[['Grades']]

plt.scatter(data['Attendance_Percentage'],data['Grades'])
plt.xlabel('Independent Variable (x)')
plt.ylabel('Dependent Variable (y)')
plt.title('Scatter Plot of Data')
plt.show()
```

# **4. Preparing Data for Modeling**

```python
x = data[['Attendance_Percentage']] #Independent
y = data[['Grades']] # Dependent

# x = data.drop('Grades', axis = 1)
# y = data.drop()
```

# **5. Fitting the Linear Regression Model**

```python
model = LinearRegression()

model.fit(x,y)
```

**6. Getting Model Coefficients**

```python
slope = model.coef_[0]

intercept = model.intercept_

print('slope: ', slope)
print('intercept:', intercept)
```

# 7. Visualizing the Linear Regression Line

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x), color='maroon')
plt.xlabel('Independent Variable (x)')
plt.ylabel('Dependent Variable (y)')
plt.title('Linear Regression')
plt.show()
```

# **8. Making Predictions**

```python
new_x = [[np.random.randint(50,100)]]

predicted_y = model.predict(new_x)

print('Value of X:', new_x)
print('Predicted Value of Y:', predicted_y)
```
