```python
from automata.fa.dfa import DFA
```

**2. Define DFA, States, Transitions, Start, Final, input symbols**

Exer2.png


```python
dfa = DFA(
    states={'q0', 'q1'},
    input_symbols={'0', '1'},
    transitions={
        'q0': {'0': 'q0', '1': 'q1'},
        'q1': {'0': 'q1', '1': 'q0'}
    },
    initial_state='q0',
    final_states={'q0'}
)
```

**3. Define the function is_accepted**

```python
def is_accepted(input_str):
  return dfa.accepts_input(input_str)
```

**4. Test the DFA**

```python
test_strings = {'110','101','000','111'}

for test_str in test_strings:
    if is_accepted(test_str):
      print(f'String "(test_str)" is accepted.')
    else:
      print(f'String "(test_str)" is not accepted.')
```
